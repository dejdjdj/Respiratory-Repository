### Caesar Cipher

### Scytale Cipher

### Vigenere Cipher

```python
message = input("What is the message? ")
key = input("What is the key? ")

def problem_3(message, key):
    
    new_key = ""
    cipher = ""
    
    # runs the for loop if key still doesn't have the same number of letters as message
    while len(message) != len(new_key):

        # prints the key to match the message length and keeps running under while
        for x in range(len(key)):  

            # if it reaches the last letter of key then it resets to the first
            if x == len(key) - 1:
                # adds a letter unless key and message are already equal
                if len(message) != len(new_key):
                    new_key += key[x] 
                    # returns x to 0 to repeat the key
                    x -= len(key)         
            else:
                # adds a letter unless key and message are already equal
                if len(message) != len(new_key):
                    new_key += key[x]
                    
   
    
    # pairs each letter in new_key and adds shift based on its counterpart in message
    for i, y in enumerate(message):
        
        # makes it so that the free spaces in the message stays that way
        if y == " ":
            cipher += " "
        else:
            # turns each letter to unicode
            unicode_k = ord(y)
            
            # turns each letter to unicode
            # -65 because it's the shift
            shift = ord(new_key[i]) - 65 

            # shifts the unicode
            new_uni_k = (unicode_k + shift)

            # if it exceeds Z it should go back to A
            if new_uni_k > 90:
                new_uni_k -= 26

            # turns back the shifted unicode into a character
            new_letter_k = chr(new_uni_k)

            # so that the output isn't just 1 letter
            cipher += new_letter_k

            
    return cipher



print(problem_3(message, key))
    
```

### Scytale De-Cipher

```python
cipher = input("What do you wish to decode? ")
shift = int(input("How many units is the shift of the message? "))


def problem_4(cipher, shift):
    
    message = ""
    
    
    i = len(cipher) // shift
    
    for y in range(shift): #[0,2] 
        
        # print rows
        for x in range(i): #[0,4] 
            message += cipher[y + (x * shift)]
            
    return message

print(problem_4(cipher, shift))
            
    
    
```
